```python
import pickle
import numpy as np
```

```python
ohe_sex = pickle.load(open('/content/drive/MyDrive/Datasets/ohe_sex.pkl', 'rb'))
ohe_Embarked = pickle.load(open('/content/drive/MyDrive/Datasets/ohe_Embarked.pkl', 'rb'))
clf = pickle.load(open('/content/drive/MyDrive/Datasets/clf.pkl','rb'))
```

```python
# Assum User Input
# Pclass | Sex | Age	| SibSp |	Parch |	Fare |	Embarked

test_input = np.array([2, 'male', 31.0, 0, 0, 10.5, 'S'], dtype=object).reshape(1,7)
test_input
```

______

```python
test_input_sex = ohe_sex.transform(test_input[:,1].reshape(1,1))
```

```python
test_input_Embarked  = ohe_Embarked.transform(test_input[:,-1].reshape(1,1))
```
